```python
import pymongo
from pymongo import MongoClient
from pymongo.errors import CursorNotFound
from pprint import pprint

MONGODB_URL = "mongodb://172.20.1.162:64102,172.20.1.163:64102/?replicaSet=configdb&readPreference=primary&ssl=false"

cursor = MongoClient(MONGODB_URL)



def find_all_services_object(db, collection ):
    """ Pesquisa documentos no mongodb"""

    try:
        documents = cursor[db][collection].find({'nome': 'instances'}, {'_id': 0, 'nome': 0})
    except CursorNotFound as err:
        raise err
    return documents


for i in find_all_services_object('__prime__', 'devops'):
    pprint(i['servers'])
```

```python
import os
from jinja2 import Environment
from jinja2 import FileSystemLoader

TEMPLATE = 'scripts'
PATH_CORTEX = 'cortex'
PREFIX = 'cs'


def create_service(name_service, output_path, template_path, port=None):
    """ Renderiza template de scrits"""
    loader = FileSystemLoader('{}'.format(template_path))
    env = Environment(loader=loader)
    template = env.get_template('csinit')
    _service = PREFIX+name_service+'-'+'2'
    output_template = os.path.join(output_path, _service)


    with open(output_template, "w") as script:
        render = template.render(name=name_service, port=port)
        script.write(render)
        script.close()


def path_services(service_name):
    list_dirs = os.listdir('../../../cortex')
    if service_name in list_dirs:
        create_service(service_name, './', os.getcwd())
        print(service_name)


path_services('operation')
```
